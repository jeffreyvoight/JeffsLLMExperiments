```python
import re
pattern = r"^(?:(?=.*[a-z])(?=.*[b])(?=.*[c])(?=.*[d])(?=.*[e])(?=.*[f])(?=.*[g])(?=.*[h])(?=.*[i])(?=.*[j])(?=.*[k])(?=.*[l])(?=.*[m])(?=.*[n])(?=.*[o])(?=.*[p])(?=.*[q])(?=.*[r])(?=.*[s])(?=.*[t])(?=.*[u])(?=.*[v])(?=.*[w])(?=.*[x])(?=.*[y])(?=.*[z])).*.*$"

def test_regex(string):
    if re.match(pattern, string, re.IGNORECASE):
        print(f"'{string}' matches")
        return True
    else:
        print(f"'{string}' does not match")
        return False

# Test a wider range of strings
test_strings = [
    "abcdefghijklmnopqrstuvwxyz",
    "zyxwvutsrqponmlkjihgfedcba",
    "zxwvutsrqponmlkjihgfedcba",
    "abc",
    "abcdefghijklmnopqrstuvwxyzzyxwvutsrqponmlkjihgfedcba", #Mix of order
    "aBcDeFgHiJkLmNoPqRsTuVwXyZ",
    "abcdefghijklmnopqrtuvwxyz", # Missing letters
    "abcdefghijklmnopqrstuvwxy", # Missing letters
    "hello world",
    "This string contains all the letters",
    "abcdefghijklmnopqrstuvwxyzabcdefghijklmnopqrstuvwxyz",
    "The quick brown fox jumps over the lazy dog.",
    "Bright stars illuminate the dark, vast night sky.",
    "Gentle breezes whisper secrets through ancient, tall trees.",
    "Golden sunlight dances upon the shimmering, cool water.",
    "Happy children laugh and play in green fields.",
    "Delicious aromas waft from the cozy, warm kitchen.",
    "Colorful birds sing sweet melodies throughout the forest.",
    "Peaceful silence descends upon the quiet, sleepy village.",
    "Mysterious shadows lengthen as the evening draws near.",
]

for s in test_strings:
    test_regex(s)
```
