```python
import matplotlib.pyplot as plt
import numpy as np

# Define the x values
x = np.linspace(-40, 90, 100)  # Create 100 points between -10 and 10

# Calculate y values for the first equation (y = x*(9/5) + 32)
y1 = (9/5)*x + 32

# Calculate y values for the second equation (y = x*2 + 30)
y2 = 2*x + 30

# Create the plot
plt.figure(figsize=(8, 6))  # Adjust figure size for better visualization

# Plot the first equation
plt.plot(x, y1, label='y1 = (9/5)x + 32', color='blue')

# Plot the second equation
plt.plot(x, y2, label='y2 = 2x + 30', color='red')

plt.plot(x, y2-y1, label='y2-y1', color='black')
# Add labels and title
plt.xlabel('x')
plt.ylabel('y')
plt.title('Comparison of y = (9/5)x + 32 and y = 2x + 30')

# Add a grid for better readability
plt.grid(True)

# Add a legend to distinguish the lines
plt.legend()

# Show the plot
plt.show()

```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the Celsius temperatures
celsius_temps = np.linspace(-40, 100, 200)

# Calculate Fahrenheit temperatures using the first formula (F = C*(9/5) + 32)
fahrenheit1 = celsius_temps * (9/5) + 32

# Calculate Fahrenheit temperatures using the second formula (F = C*2 + 30)
fahrenheit2 = celsius_temps * 2 + 30

# Calculate the difference between the two formulas
difference = fahrenheit1 - fahrenheit2

# --- First Graph: F vs. C ---
plt.figure(figsize=(10, 6))
plt.plot(celsius_temps, fahrenheit1, label='F = C*(9/5) + 32', color='blue')
plt.plot(celsius_temps, fahrenheit2, label='F = C*2 + 30', color='red')
plt.plot(celsius_temps, difference, label='Difference (F1 - F2)', color='green',
linestyle='--')
plt.xlabel('Celsius (C)')
plt.ylabel('Fahrenheit (F)')
plt.title('Comparison of Fahrenheit Conversions (F vs. C)')
plt.grid(True)
plt.legend()

# --- Second Graph: C vs. F (Inverse) ---
plt.figure(figsize=(10, 6))

# Inverse of F1: C1 = (F - 32) * (5/9)
celsius1 = (fahrenheit1 - 32) * (5/9)

# Inverse of F2: C2 = (F - 30) / 2
celsius2 = (fahrenheit2 - 30) / 2

plt.plot(fahrenheit1, celsius1, label='C = (F - 32) * (5/9)', color='blue')
plt.plot(fahrenheit2, celsius2, label='C = (F - 30) / 2', color='red')
plt.xlabel('Fahrenheit (F)')
plt.ylabel('Celsius (C)')
plt.title('Conversion from Fahrenheit to Celsius (F to C)')
plt.grid(True)
plt.legend()

plt.show()
```
